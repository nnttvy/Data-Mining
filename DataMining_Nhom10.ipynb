# **Apriori**

```python
!pip install -q mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

transactions = df.fillna('').apply(lambda x: set(x), axis=1).tolist()
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
transformed_data = pd.DataFrame(te_ary, columns=te.columns_)
transformed_data = transformed_data.drop(columns=transformed_data.columns[0])
```

```python
#Áp dụng thuật toán apriori tìm frequent itemsets với minSup = 0.03
frequent_itemsets_apriori = apriori(transformed_data, min_support=0.03, use_colnames=True)

#Hiển thị các itemsets có từ 2 items trở lên và theo thứ tự support giảm dần
frequent_itemsets_apriori[frequent_itemsets_apriori['itemsets'].apply(lambda x: len(x) >= 2)].sort_values(['support'], ascending = [False])
```

```python
from mlxtend.frequent_patterns import association_rules

# Tìm luật mạnh từ frequent itemsets với minConf = 0.4
rules_apriori = association_rules(frequent_itemsets_apriori, metric = "confidence", min_threshold = 0.4)
rules_apriori = rules_apriori.sort_values(['support'], ascending = [False])

# Chỉ in các chỉ số cần thiết
rules_apriori[['antecedents', 'consequents','antecedent support',
              'consequent support', 'support', 'confidence', 'lift']]
```

# **ECLAT**

```python
!pip install pyECLAT
```

```python
from pyECLAT import ECLAT
eclat = ECLAT(data=df)

# Chuyển về binary dataframe
binary_df = eclat.df_bin
print("Binary DataFrame:")
print(binary_df[:5])
```

```python
#Xem số lượng bán ra của các món hàng
items_total = eclat.df_bin.astype(int).sum(axis=0)
items_total
```

```python
#Xem số lượng món hàng được mua bởi mỗi transaction
items_per_transaction = eclat.df_bin.astype(int).sum(axis=1)
items_per_transaction
```

```python
#Áp dụng thuật toán ECLAT tìm frequent itemsets với minSup = 0.03
get_ECLAT_indexes, get_ECLAT_supports = eclat.fit_all(min_support=0.03,
                                                      min_combination=1, separator=', ')
```

```python
# Tạo dataframe chứa frequent itemsets
frequent_itemsets_eclat = pd.DataFrame({'support': get_ECLAT_supports.values(),'itemsets': get_ECLAT_supports.keys()})
frequent_itemsets_eclat = frequent_itemsets_eclat.sort_values(by='support', ascending=False)
frequent_itemsets_eclat
```

```python
#Chuyển itemsets về kiểu frozenset để phù hợp với format của hàm association_rule()
frequent_itemsets_eclat['itemsets'] = frequent_itemsets_eclat['itemsets'].apply(lambda x: frozenset(x.split(', ')))
# frequent_itemsets_eclat['itemsets'] = tuple(frequent_itemsets_eclat['itemsets'])
# frequent_itemsets_eclat['itemsets'] =frequent_itemsets_eclat['itemsets'].apply(frozenset)
frequent_itemsets_eclat
```

```python
from mlxtend.frequent_patterns import association_rules

# Tạo bảng luật kết hợp
rules_eclat = association_rules(frequent_itemsets_eclat, metric = "support",  min_threshold = 0.03
              ,support_only=True)
rules_eclat = rules_eclat.sort_values(['support'], ascending = [False])

# Chỉ in các chỉ số cần thiết
'''
rules_eclat[['antecedents', 'consequents','antecedent support',
       'consequent support', 'support', 'confidence', 'lift','conviction']]
'''

rules_eclat[['antecedents', 'consequents','support']]
```

```python
#Tính chỉ số confidence cho từng luật
conf_list = []
for i, row in rules_eclat.iterrows():
  conf = row['support']*len(df)/int(binary_df[row['antecedents']].sum()[0])
  conf_list.append(conf)
rules_eclat['confidence'] = conf_list
```

```python
#In những luật kết hợp mạnh, minConf = 0.4
rules_eclat=rules_eclat[rules_eclat['confidence']>=0.4][['antecedents', 'consequents','support', 'confidence']]
rules_eclat
```

# **Đánh giá**

## Độ đo Lift

```python
lift_list = []
for i, row in rules_eclat.iterrows():
  lift = row['confidence']/(int(binary_df[row['consequents']].sum()[0])/len(df))
  lift_list.append(lift)

rules_eclat['lift'] = lift_list
```

```python
#X, Y độc lập
rules_eclat[rules_eclat.lift == 1][['antecedents', 'consequents','support', 'confidence', 'lift']]
```

```python
#X, Y tương quan thuận
rules_eclat[rules_eclat.lift > 1][['antecedents', 'consequents','support', 'confidence', 'lift']]
```

```python
#X, Y tương quan nghịch
rules_eclat[rules_eclat.lift < 1][['antecedents', 'consequents','support', 'confidence', 'lift']]
```

## Độ đo Chi-square

```python
from scipy.stats import chi2_contingency

chi2_list = []
expected_freq = []
freq = []
p_value = []

for i, row in rules_eclat.iterrows():

  # số lần xuất hiện đồng thời của antecedents và consequents
  ob1 = row['support']*len(df)

  # số lần antecedents xuất hiện nhưng consequents không xuất hiện
  ob2 = int(binary_df[row['antecedents']].sum()[0]) - ob1

  # số lần consequents xuất hiện nhưng antecedents không xuất hiện
  ob3 = int(binary_df[row['consequents']].sum()[0]) - ob1

  # số lần cả 2 đều không xuất hiện
  ob4 = (len(df)-int(binary_df[row['antecedents']].sum()[0])) - ob3

  # tạo contingency table
  table = np.array([[ob1,ob2],[ob3,ob4]])
  contingency_table = chi2_contingency(table)

  chi2 = contingency_table.statistic
  chi2_list.append(round(chi2))
  p_value.append(contingency_table.pvalue)
  expected_freq.append(round(contingency_table.expected_freq[0,0])) # lấy giá trị kỳ vọng cả 2 cùng xuất hiện
  freq.append(round(ob1))

rules_eclat['Chi2'] = chi2_list
rules_eclat['pvalue'] = p_value
rules_eclat['freq'] = freq
rules_eclat['expected_freq'] = expected_freq
rules_eclat[['antecedents', 'consequents','Chi2','pvalue','freq','expected_freq']]
```
